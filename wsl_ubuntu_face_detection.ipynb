```python
import cv2
import sys

cap = cv2.VideoCapture(0, cv2.CAP_V4L2)
cap.set(cv2.CAP_PROP_FOURCC, cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'))

# Tensorflow model
model = './face_detector/opencv_face_detector_uint8.pb'
config = './face_detector/opencv_face_detector.pbtxt.txt'

face_net = cv2.dnn.readNet(model, config)

if face_net.empty():
    print('model load failed')
    sys.exit()

h, w = cap.get(4), cap.get(3)
cap.set(4, h * 2)
cap.set(3, w * 2)
h, w = cap.get(4), cap.get(3)

while True:
    ret, frame = cap.read()

    if ret == False:
        print('image read failed')
        break
    
    blob = cv2.dnn.blobFromImage(frame, 1, (300, 300), (104, 177, 123), swapRB=False)
    face_net.setInput(blob)
    out = face_net.forward()
    output = out[0, 0, :, :]

    for i in range(output.shape[0]):
        confidence = output[i, 2]
        if confidence > 0.5:
            x1 = int(output[i, 3] * w)
            y1 = int(output[i, 4] * h)
            x2 = int(output[i, 5] * w)
            y2 = int(output[i, 6] * h)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2, cv2.LINE_AA)
            text = f'Face: {confidence * 100:.2f}%'
            cv2.putText(frame, text, (x1, y1 - 5), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
        
    cv2.imshow('frame', frame)
    key = cv2.waitKey(50)
    if key == 27:
        break
    if key == ord(' '): # 일시정지 ㅋ
        cv2.waitKey()

cap.release()
cv2.destroyAllWindows()
```
